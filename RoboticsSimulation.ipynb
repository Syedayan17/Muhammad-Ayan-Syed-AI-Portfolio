# Robotics Simulation
Simple movement simulation using coordinates.

```python
position = [0,0]
moves = ['UP','UP','RIGHT','DOWN','LEFT']
```

```python
for move in moves:
    if move == 'UP': position[1]+=1
    elif move == 'DOWN': position[1]-=1
    elif move == 'RIGHT': position[0]+=1
    elif move == 'LEFT': position[0]-=1
```

```python
print('Final Position:', position)
```
